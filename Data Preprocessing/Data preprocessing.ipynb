# To replace missing value function in pandas is df[column_name].fillna(value)

```python
def replacer(df):
    cat = list(df.columns[df.dtypes == object])
    for col in df.columns:
        if col in cat:
            md = df[col].mode()[0]
            df[col]=df[col].fillna(md)
        else:
            mn = df[col].mean()
            df[col] = df[col].fillna(mn)
    print('Missing values Replaced')
```

```python
replacer(df)
```

# Check mse, rmse, mae, r2 , mape

```python
def evaluate_model (model , x , y):
    ### predicting the model for x
    y_pred  = model.predict(x)
    ###Evaluating the model using necessary packages
    from sklearn.metrics import mean_squared_error , mean_absolute_error
    from sklearn.metrics import r2_score, mean_absolute_percentage_error
    mse = mean_squared_error(y , y_pred)
    mae = mean_absolute_error(y , y_pred)
    rmse = mse*(1/2)
    mape = mean_absolute_percentage_error(y , y_pred)
    r2_score = r2_score(y , y_pred)
    ### printing the above evaluating models
    print(f'Mean squared Error :{mse:.2f}')
    print(f'Mean absolute Error :{mae:.2f}')
    print(f'Root mean squared Error :{rmse:.2f}')
    print(f'Mean absolute percentage Error :{mape:.2f}')
    print(f'R2_score :{r2_score:.2f}')
```

```python
evaluate_model(model , x_train , y_train)
```

```python
evaluate_model(model , x_test , y_test)
```
