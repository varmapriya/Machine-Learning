# 5. Sklearn Pipeline  1. Feature selection (Ordinal Encode) 

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder , StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute' , SimpleImputer(strategy='mean')),
                            ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute' , SimpleImputer(strategy='constant',fill_value='Not_Avail')),
                            ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num' ,num_pipe1, con),
                          ('cat', cat_pipe1 , cat)]).set_output(transform='pandas')
```

```python
Xpre = pre1.fit_transform(X)
Xpre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr , n_features_to_select='auto',direction='forward')
sel.fit_transform(Xpre , Y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
(sel_cols)[0]
```

```python
sel_cols[0].split('__')
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)
```

```python
imp_cols
```

```python
len(imp_cols)
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

# 5. Sklearn Pipeline   2. Final Pipeline(OneHotEncoder)

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
cat_sel
```

```python
con
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='constant' , fill_value='Not_Avail')),
                            ('ohe',OneHotEncoder(handle_unknown='ignore' , sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num' , num_pipe2 , con_sel),
                          ('cat' , cat_pipe2 , cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
X_sel_pre = pre2 . fit_transform(X_sel)
X_sel_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel2 = SequentialFeatureSelector(estimator=lr , n_features_to_select='auto',direction='backward')
sel2.fit_transform(X_sel_pre , Y)
sel2_cols = sel2.get_feature_names_out()
sel2_cols
```
