# Multiple Linear Regression

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

# Reading the Dataset

```python
import pandas as pd
df = pd.read_csv('50_Startups.csv')
df.head()
```

# PROFIT(Y) ~ RND(X1), ADMIN(X2), MKT(X3)
# PROFIT = B0 + B1xRND + B2xADMIN + B3xMKT

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

# Seperating X and Y

```python
X = df[['RND', 'ADMIN', 'MKT']]#####independent value
Y = df[['PROFIT']]#####dependent value
```

```python
X.head()
```

```python
Y.head()
```

# Creating a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.coef_ ###means 'm' i.e slope
```

```python
model.intercept_
```

```python
X.columns
```

# PROFIT_pred = 50122.19 + 0.8057xRND - 0.0268xADMIN + 0.0272xMKT

# To print the above coeff model in series

```python
coef = pd.Series(model.coef_.flatten())
coef.index = X.columns ###to print the index value as the names of columns
coef
```

```python
coef.plot(kind='bar',title  ='Coefficient of Linear Regression')
```

# 1.IF RND increases by 1000 , PROFIT increases by 805.7
# 2.IF ADMIN increase by 1000 , PROFIT decrease by 26.8
# 3.IF MKT increases by 1000, PROFIT increase by 27.2

# Evaluate the Model

# MSE, RMSE, MAE , MAPE ,R2 Score

```python
model.score(X,Y)
```

```python
Y_pred = model.predict(X)
Y_pred[0:5]
```

```python
Y.head()
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error
mse = mean_squared_error(Y,Y_pred)
mse
```

```python
mae = mean_absolute_error(Y,Y_pred)
mae
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(Y,Y_pred)
mape
```

```python
## for RMSE
rsme = mse**(1/2)
rsme
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(Y,Y_pred)
r2_score
```

# Above model is good and can be used out of sample predictions
