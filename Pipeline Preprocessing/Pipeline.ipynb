
# Step 3 : Seperate X and Y

```python
X = df.drop(columns=['id','Weight'])
Y = df[['Weight']]
```

```python
Y.head()
```

# Step 4 : Create a preprocessing pipeline on X

```python
X.columns
```

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes =='object'])
con = list(X.columns[X.dtypes !='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline # Used to automate preprocessing
from sklearn.impute import SimpleImputer # Used to replace missing values
from sklearn.preprocessing import StandardScaler,OneHotEncoder # Preprocessing
from sklearn.compose import ColumnTransformer # Used to combine cat and con Pipeline
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer (strategy = 'mean')),
                           ('scaler' , StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe',OneHotEncoder(handle_unknown='ignore' , sparse_output=False))])
```

```python
pre = ColumnTransformer([('num' , num_pipe , con),
                         ('cat' , cat_pipe , cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
X_pre.to_csv('Processedcsv.csv' , index=False)
```

# Step 5: Apply Train test split
### 20% of data will be unseen to model

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test =train_test_split(X_pre , Y , random_state= 21 , test_size=0.2)
```

# Step 6 : Built the ML model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

# Step 7 : Evaluate the regression mode

#### mse , mae , mape , r2 score , rmse

```python
def evaluation_model(model , x ,y):
    ##predicting the model for x
    y_pred = model.predict(x)
    ### importing necessary libraries
    from sklearn.metrics import mean_absolute_error , mean_absolute_percentage_error ,mean_squared_error,r2_score 
    mse = mean_squared_error(y,y_pred)
    rmse = mse*(1/2)
    mae = mean_absolute_error(y,y_pred)
    mape = mean_absolute_percentage_error(y ,y_pred)
    r2_score = r2_score(y,y_pred)
    ###printing above models
    print(f'mean squared error: {mse:.2f}')
    print(f'root mean squared error : {rmse:.2f}')
    print(f'mean absolute error: {mse:.2f}')
    print(f'mean absolute percentage error :{mape :.2f}')
    print(f'r2_score : {r2_score:.2f}')
```

```python
evaluation_model
```

```python
evaluation_model(model, x_train, y_train)
```

```python
evaluation_model(model, x_test , y_test)
```
